```python
base_url = 'https://dvago.pk'
product_urls_list = []

for i in range(1,150):
    url = f'https://dvago.pk/collections/medicine?page={i}&grid_list=grid-view'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    div_tags = (soup.find_all('div', {'class':'productitem'}))
    for item in div_tags:
        product_urls_list.append(base_url+(item.find('a')['href']))
```

```python
def product_details(url):
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    product_detail = dict()
    
    try:
        product_detail['Title'] = soup.find('h1', {'class':'product-title'}).text.strip()
    except:
        product_detail['Title'] = None
    # print(title)
    
    try:
        product_detail['Vendor'] = soup.find('div',{'class':'product-vendor'}).find('a').text
    except:
        product_detail['Vendor'] = None
    # print(vendor)
    
    try:
        product_detail['Price'] = soup.find('div', {'class':'price--main'}).text.strip()
    except:
        product_detail['Price'] = None
        
    try:
        product_detail['Unit price'] = soup.find('span', {'id':'price-per-unit'}).text
    except:
        product_detail['Unit price'] = None
    # print(unit_price)
    
    try:
        product_detail['Total units'] = soup.find('input', {'id':'units-per-case'}).get('value')
    except:
        product_detail['Total units'] = None
    # print(total_units)
    
    try:
        product_detail['Category'] = soup.find_all('div',{'class':'tit'})[1].find('a').text
    except:
        product_detail['Category'] = None
    # print(category)
    
    try:
        overview_heads = [head.text for head in soup.find('div', {'id':'tab-1'}).find_all('div', {'class':'head'})]
    # print(heads)
        overview = [overview.text for overview in soup.find('div', {'id':'tab-1'}).find_all('div', {'class':'text'})]
    # print(overview)
        product_detail['Overview'] = dict(zip(overview_heads,overview))
    except:
        product_detail['Overview'] = None

    try:
        warning_heads = [head.text for head in soup.find('div', {'id':'tab-2'}).find_all('div', {'class':'head'})][3:]
        # print(warning_heads)
        warning = [overview.text for overview in soup.find('div', {'id':'tab-2'}).find_all('div', {'class':'text'})]
        warning = [warning[:4]] + warning[4:]
        # print(warning)
        product_detail['Warnings'] = dict(zip((warning_heads), warning))
    except:
        product_detail['Warnings'] = None

    return product_detail
```

```python
url = 'https://dvago.pk/collections/medicine/products/eziday_25mg_1x20s'
print(product_details(url))
```

```python
all_product_details = []
for url in product_urls_list:
    print(url)
    all_product_details.append(product_details(url))
```
